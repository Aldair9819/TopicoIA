```python
import face_recognition
from sklearn.neighbors import KNeighborsClassifier
import os


```

```python
# Directorio que contiene las im√°genes de las caras de entrenamiento
base_dir = 'C:\\Users\\Waldosir\\Documents\\2doCodigo\\TopicoIA\\Marzo\\BD\\corpus_images_chat\\Train\\'

# Listas para almacenar las caras de entrenamiento y las etiquetas correspondientes
images  = []
labels  = []

```

```python
# Iterar sobre cada directorio en el directorio base
for label in os.listdir(base_dir):
    label_dir = os.path.join(base_dir, label)
    if os.path.isdir(label_dir):
        # Obtener las im√°genes en el directorio y sus etiquetas
        for filename in os.listdir(label_dir):
            print(filename)
            try:
                image = face_recognition.load_image_file(os.path.join(label_dir, filename))
                face_encoding = face_recognition.face_encodings(image)[0]
                images.append(face_encoding)
                labels.append(label)
            except Exception as e:
                print('No se pudo cargar la imagen', filename, 'Error:', e)
```

```python
print(images)
```

```python
# Crear y entrenar el clasificador KNN
knn_classifier = KNeighborsClassifier(n_neighbors=3)  # Puedes ajustar el n√∫mero de vecinos seg√∫n sea necesario
knn_classifier.fit(images, labels)


# Directorio que contiene las im√°genes de las caras de prueba
test_dir = 'C:\\Users\\Waldosir\\Documents\\2doCodigo\\TopicoIA\\Marzo\\BD\\corpus_images_chat\\Test\\'

```

```python
# Predecir las etiquetas de las im√°genes de prueba
# Iterar sobre cada directorio en el directorio base



for label in os.listdir(test_dir):
    label_dir = os.path.join(test_dir, label)
    if os.path.isdir(label_dir):
        # Obtener las im√°genes en el directorio y sus etiquetas
        for filename in os.listdir(label_dir):
            print(filename)
            try:
                image_path = os.path.join(test_dir, filename)
                image = face_recognition.load_image_file(os.path.join(label_dir, filename))
                face_encodings = face_recognition.face_encodings(image)
                predicted_label = knn_classifier.predict([face_encodings[0]])
                print(f'La imagen {filename} parece ser: {predicted_label[0]} y tiene la ruta {label_dir}')
            except Exception as e:
                print('No se pudo cargar la imagen', filename, 'Error:', e)

```
